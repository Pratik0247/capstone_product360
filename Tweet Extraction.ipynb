```
text = "nokia 6.1 plus"
count = 4000

tweetCriteria2 = got.manager.TweetCriteria().setQuerySearch(text)\
                                                    .setMaxTweets(count)
```

```
tweets2 = got.manager.TweetManager.getTweets(tweetCriteria2)

text_tweets2 = [[tw.text, tw.hashtags] for tw in tweets2]
```
